```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
import seaborn as sns
sns.set(color_codes=True)
import datetime
```

```python
dt=pd.read_excel("/content/Housing.xlsx")
dt
```

Replace the missing values with appropriate approximations

```python
dt['Acre'].fillna(np.mean(dt["Acre"]),inplace=True)
dt['LandValue'].fillna(np.mean(dt["LandValue"]),inplace=True)
dt['BuildingValue'].fillna(np.mean(dt["BuildingValue"]),inplace=True)
dt['TotalValue'].fillna(np.mean(dt["TotalValue"]),inplace=True)
dt['YearBuilt'].fillna(np.mean(dt["YearBuilt"]),inplace=True)
dt['Bedrooms'].fillna(np.mean(dt["Bedrooms"]),inplace=True)
dt['FullBath'].fillna(np.mean(dt["FullBath"]),inplace=True)
dt['HalfBath'].fillna(np.mean(dt["HalfBath"]),inplace=True)
dt
```

```python
dt.isna()
```

Rename the columns to better suited names

```python
dt.rename(columns={"PropertyAddress":"PropAddr"},inplace=True)
dt.rename(columns={"SalePrice":"SalesPrice"},inplace=True)
dt.rename(columns={"Bedrooms":"Number_of_Bedrooms"},inplace=True)
dt
```

Extract the month and year from the sale date.

```python
dt['month'] = dt['SaleDate'].dt.month
dt['month']
```

```python
dt['year'] = dt['SaleDate'].dt.year
dt['year']
dt
```

Extract the pin number from the property address

```python
dt['pin']=dt['PropAddr'].str[:4]
dt['pin']
```

Use appropriate plots and functions to answer the following questions.

1.How has the price of properties changed over the years

```python
c=dt.groupby("year")["SalesPrice"].count()
sns.barplot(x=c.index,y=c.values)
```

2.Does a specific month or span of months have the highest property prices?

```python
c=dt.groupby("month")["SalesPrice"].count()
sns.barplot(x=c.index,y=c.values)
```

3.For what purpose were the properties generally purchased

```python
c=dt.groupby("LandUse")["LandUse"].count()
c
```

```python
sns.barplot(x=c.index,y=c.values)
```

Have the age of houses affected the price?

```python
d=dt.groupby("age")["SalesPrice"].count()
d
sns.scatterplot(x=d.index,y=d.values)
```
